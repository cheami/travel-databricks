# Create and load data


```sql
-- 1. Create the target table in your travel schema
CREATE TABLE IF NOT EXISTS workspace.travel.transactions;
CREATE TABLE IF NOT EXISTS workspace.travel.manual_logs;
```

```sql
-- 2. Copy the data from the Volume into the table
COPY INTO workspace.travel.transactions
FROM '/Volumes/workspace/travel/travel_data_volume/transactions'
FILEFORMAT = CSV
FORMAT_OPTIONS (
  'header' = 'true',
  'inferSchema' = 'true'
)
COPY_OPTIONS (
  'mergeSchema' = 'true'
);

COPY INTO workspace.travel.manual_logs
FROM '/Volumes/workspace/travel/travel_data_volume/manual_logs'
FILEFORMAT = CSV
FORMAT_OPTIONS (
  'header' = 'true',
  'inferSchema' = 'true'
)
COPY_OPTIONS (
  'mergeSchema' = 'true'
);
```
